```python
import pandas as pd
import numpy as np
```

```python
best = []
best_avg = []
best_std = []
fitness_avg = []
fitness_std = []
time_avg = []
time_std = []
it_avg = []
it_std = []
size_avg = []
size_std = []
access_avg = []
access_std = []

paths = ['solutions-new/wines/grasp/solution', 'solutions-new/wines/ils/solution', 'solutions-new/wines/ga/solution', 'solutions-new/wines/vns/solution', 'solutions-new/wines/vns-inv/solution']

for l, x in enumerate(paths):
    all_times = []
    all_fitness = []
    all_best = []
    all_it_best = []
    all_time_best = []
    all_hash_size = []
    all_hash_access = []
  
    for i in range(0, 10):
        data = pd.read_csv(x + str(i+1) +  '.csv')
        for j, rows in enumerate(data.values):
            if j == 1:
                all_times.append(rows[0].split(";")[0])
            elif j >= 6 and j <= 15:
                if j == 6:
                    all_best.append(rows[0].split(";")[0])
                all_fitness.append(rows[0].split(";")[0])
            elif j == 17:
                all_it_best.append(rows[0].split(";")[3])
                all_time_best.append(rows[0].split(";")[4])
                all_hash_size.append(rows[0].split(";")[5])
                all_hash_access.append(rows[0].split(";")[6])
    
    all_best = [float(x) for x in all_best]
    all_fitness = [float(x) for x in all_fitness]
    all_time_best = [float(x) for x in all_time_best]
    all_it_best = [float(x) for x in all_it_best] 
    all_hash_size = [float(x) for x in all_hash_size]
    all_hash_access = [float(x) for x in all_hash_access]

    all_best.sort(reverse=True)
    
    best.append(all_best[0])
    best_avg.append(round(np.mean(all_best), 6))  
    best_std.append(round(np.std(all_best), 6))
    fitness_avg.append(round(np.mean(all_fitness), 6))
    fitness_std.append(round(np.std(all_fitness), 6))
    time_avg.append(round(np.mean(all_time_best), 6)) 
    time_std.append(round(np.std(all_time_best), 6))
    it_avg.append(round(np.mean(all_it_best), 6))
    it_std.append(round(np.std(all_it_best), 6))
    size_avg.append(round(np.mean(all_hash_size), 6))
    size_std.append(round(np.std(all_hash_size), 6))
    access_avg.append(round(np.mean(all_hash_access), 6))
    access_std.append(round(np.std(all_hash_access), 6))
    
    print('')
    print('Path: ', x)
    print('')
    #print("Times: ", all_times)
    #print("all_fitness", all_fitness)
    #print("all_best", all_best)
    #print("all_it_best", all_it_best)
    #print("all_time_best", all_time_best)
    #print("all_hash_size", all_hash_size)
    #print("all_hash_access", all_hash_access)
    
    print('Best: ', best[l])
    print('Best avg: ', best_avg[l])
    print('Best std: ', best_std[l])
    print('All fitness avg: ', fitness_avg[l])
    print('All fitness std: ', fitness_std[l])
    print('Time avg: ', time_avg[l])
    print('Time std: ', time_std[l])
    print('it_avg: ', it_avg[l])
    print('it_std: ', it_std[l])
    print('size_avg: ', size_avg[l])
    print('size_std: ', size_std[l])
    print('access_avg: ', access_avg[l])
    print('access_std: ', access_std[l])

```
